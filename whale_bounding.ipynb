```python
import os

import PIL
from PIL import Image
from PIL.ImageDraw import Draw
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from keras.models import load_model
from keras.preprocessing import image
```

```python
MODEL_BASE = '../bounding/'
DATA = '../dados/'
CROP_TRAIN_DATA = '../bounding/crop_train/'
CROP_TEST_DATA = '../bounding/crop_test/'
TRAIN_DATA = os.path.join(DATA, 'train')
TEST_DATA = os.path.join(DATA, 'test')
```

```python
model = load_model(os.path.join(MODEL_BASE, 'cropping_my.model'))
#model.summary()
```

```python
train_paths = [os.path.join(TRAIN_DATA, img) for img in os.listdir(TRAIN_DATA)]
test_paths = [os.path.join(TEST_DATA, img) for img in os.listdir(TEST_DATA)]
```

```python
def make_bbox_image(img_path):
    """
    :param img: path to image
    """
    main_img = image.load_img(img_path)
    # Original size
    origin_a, origin_b = main_img.size
    
    r_img = main_img.resize((128, 128), PIL.Image.ANTIALIAS)
    # convert to 1d image
    rb_img = r_img.convert('L')
    rb_img_arr = image.img_to_array(rb_img)
    bbox = model.predict(np.expand_dims(rb_img_arr, axis=0))
    print(bbox[0])
    
    # draw rectangle
    # draw = Draw(rimg)
    # draw.rectangle(bbox, outline='red')
    var_a = origin_a / 128
    var_b = origin_b / 128
    
    a,b,c,d = bbox[0]
    
    a = a * var_a
    b = b * var_b
    c = c * var_a
    d = d * var_b
    if(a < 0):
        c = origin_a
        a = 0
    if(b < 0):
        d = origin_b
        b = 0
    if(c < 0):
        a = origin_a
        c = 0
    if(d < 0):
        b = origin_b
        d = 0
 
    
    #img_crop = r_img.crop(tuple(bbox[0]))
    #print(np.array([a,b,c,d]))
    img_crop = main_img.crop(tuple(np.array([a,b,c,d])))
    #img_arr = image.img_to_array(img_crop)
    return img_crop
```

```python
#print(train_paths[1])
#img = make_bbox_image(os.path.join(TRAIN_DATA, '0041880bf.jpg'))
#img = make_bbox_image(train_paths[1])
#img

count = 0
limit = 0
exp_list = []

import cv2

for img in os.listdir(TRAIN_DATA):
    #if(limit == 200):
    #    break
    #limit += 1
    try:
        path = os.path.join(TRAIN_DATA, img)
        f_img = make_bbox_image(path)
        #t = np.asarray(f_img)
        #cv2.imwrite(CROP_TRAIN_DATA + str(img), t)
        f_img.save(CROP_TRAIN_DATA + str(img))
    except:
        a = cv2.imread(path)
        cv2.imwrite(CROP_TRAIN_DATA + str(img), a)
        count += 1
        exp_list.append(img)
        
    
for img in os.listdir(TEST_DATA):
    try:
        path = os.path.join(TEST_DATA, img)
        f_img = make_bbox_image(path)
        #cv2.imwrite(CROP_TEST_DATA + str(img), f_img)
        f_img.save(CROP_TEST_DATA + str(img))
    except:
        a = cv2.imread(path)
        cv2.imwrite(CROP_TEST_DATA + str(img), a)
        count += 1
        exp_list.append(img)
        
print(count)
```
